Scaling ``tags`` column in range (0,1) (TODO)

```python
train_ft = train.drop(['url','class', 'tags'], axis=1)
train_ft.head()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=0).fit(train_ft) #because we found two templates
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

X_std = StandardScaler().fit_transform(train_ft)
reduced_data = PCA(n_components=2).fit_transform(X_std)
kmeans_2 = KMeans(n_clusters=3, random_state=0)
kmeans_2.fit(reduced_data)
```

```python
centroids = kmeans_2.cluster_centers_
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
plt.scatter(reduced_data[kmeans_2.labels_ == 0, 0], reduced_data[kmeans_2.labels_ == 0, 1],
            c='green', label='cluster 1')
plt.scatter(reduced_data[kmeans_2.labels_ == 1, 0], reduced_data[kmeans_2.labels_ == 1, 1],
            c='blue', label='cluster 2')
plt.scatter(reduced_data[kmeans_2.labels_ == 2, 0], reduced_data[kmeans_2.labels_ == 2, 1],
            c='yellow', label='cluster 3')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=300,
            c='r', label='centroid')

plt.legend()
plt.title('Visualization of clustered data', fontweight='bold')
ax.set_aspect('equal');
```

```python
y_test=test['class']
test_ft = test.drop(['url','class','tags'], axis=1)
test_ft.head()
```

```python
y_predicted=kmeans.predict(test_ft)
```

```python
errors=y_predicted-y_test
errors
```

```python
def precision(arr,tot):
    corrette=0
    for i in arr:
        if i==0:
            corrette=corrette+1
    return corrette/tot
```

```python
precision(errors,len(y_test))
```
